```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/content/profiles_revised.csv"
clean_data = pd.read_csv(file_path)
fig, axes = plt.subplots(3, 2, figsize=(14, 12))
sns.histplot(clean_data['age'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')
sns.countplot(y=clean_data['body_type'], order=clean_data['body_type'].value_counts().index, ax=axes[0, 1])
axes[0, 1].set_title('Body Type Distribution')
sns.countplot(x=clean_data['sex'], ax=axes[1, 0])
axes[1, 0].set_title('Gender Distribution')
sns.countplot(y=clean_data['education'], order=clean_data['education'].value_counts().index, ax=axes[1, 1])
axes[1, 1].set_title('Education Level Distribution')
sns.histplot(clean_data['income'], kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Income Distribution')
sns.countplot(y=clean_data['religion'], order=clean_data['religion'].value_counts().nlargest(5).index, ax=axes[2, 1])
axes[2, 1].set_title('Top 5 Religions Distribution')
plt.tight_layout()
plt.show()
```
