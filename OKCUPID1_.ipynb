```python



import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = {
    'religion': ['other', 'christianity', 'catholicism', 'buddhism', 'judaism', 'missing', 'islam', 'agnosticism', 'atheism', 'hinduism'],
    'stem': [500, 800, 600, 200, 100, 400, 50, 300, 250, 50],
    'other': [5500, 7500, 6500, 3500, 2500, 9500, 500, 8500, 6000, 400]
}

df = pd.DataFrame(data)
df_melt=df.melt(id_vars='religion', var_name='Class', value_name='count')
plt.figure(figsize=(10, 6))
sns.barplot(x='religion', y='count', hue='Class', data=df_melt, palette={'stem': 'lightblue', 'other': 'blue'})
plt.xlabel('')
plt.ylabel('Count')
plt.title('Count of STEM and Non-STEM Profiles by Religion')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = {
    'religion': ['other', 'christianity', 'catholicism', 'buddhism', 'judaism', 'missing', 'islam', 'agnosticism', 'atheism', 'hinduism'],
    'stem': [500, 800, 600, 200, 100, 400, 50, 300, 250, 50],
    'other': [5500, 7500, 6500, 3500, 2500, 9500, 500, 8500, 6000, 400]
}

df = pd.DataFrame(data)
df['stem_proportion'] = df['stem'] / (df['stem'] + df['other'])
df['other_proportion'] = df['other'] / (df['stem'] + df['other'])

plt.figure(figsize=(10, 6))
df.plot(kind='bar', x='religion', y=['stem_proportion', 'other_proportion'], stacked=True, color=['lightblue', 'darkblue'], ax=plt.gca())

# Labels and title
plt.xlabel('')
plt.ylabel('Proportion')
plt.title('Proportion of STEM and Non-STEM Profiles by Religion')
plt.legend(['STEM', 'Other'], loc='upper left')

plt.show()
```
